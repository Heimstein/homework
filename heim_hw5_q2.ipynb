```r
dataTemp<-read.csv('auto-mpg.data',header=FALSE,stringsAsFactors=T)
data<-na.omit(dataTemp)
names(data) <- c('mpg','cylinders','displacement','horsepower','weight','acceleration','model_year','origin','name')
data$cylinders<-data$displacement<-data$model_year<-data$origin<-data$name<-NULL
#get rid of unneccesary columns
print(data)
```

```r
library(caret)
library(datasets)
split<-createDataPartition(y=data$acceleration,p = 0.8,list=FALSE)
trainingData<-data[split,]
testingData<-data[-split,]#split data


trainingData$horsepower <- as.numeric(trainingData$horsepower)
trainingData$acceleration <- as.numeric(trainingData$acceleration)
trainingData$weight <- as.numeric(trainingData$weight)
trainingData$mpg <- as.numeric(trainingData$mpg)#change data to numeric
```

```r
lmFit<-lm(mpg~ horsepower+weight+acceleration, data=trainingData)#compute model
```

```r
summary(lmFit)
```

```r
testingData$horsepower <- as.numeric(testingData$horsepower)
testingData$acceleration <- as.numeric(testingData$acceleration)
testingData$weight <- as.numeric(testingData$weight)
testingData$mpg <- as.numeric(testingData$mpg) #change data to numeric


p<-predict(lmFit,testingData) #predicted results
p
```

```r
trueResults<-testingData$mpg# actual results
trueResults
```

```r
sum((p-trueResults)**2)  #compute squared sum of errors
```
