```r
air = read.csv(file="AirQualityUCI.csv", sep=";")
air1 <- subset(air, select = -c(X) )
air <- subset(air1,select=-c(X.1))
air<-na.omit(air)
air$T <- gsub(',', '.', air$T)
```

```r
meanSequence <- function(x){
    incrementValue=1
    result=0
    dataSet=0 # is eventual gotten rid of
    for (i in x){
        result=result+(1/incrementValue)*(i-result)
        dataSet<- c(dataSet, result)
        incrementValue=incrementValue+1
    }
    return (dataSet[-1]) # get rid of initial 0
}
```

```r
data <- as.numeric(air$T)  # by column name
#data
result<-meanSequence(data)
print(mean(data))
print(result[9357])
```

```r
result<-round(result, digits=2)
data<-round(data,digits=2)
plot(data,col="red",main="Data set and Running Average", xlab="index", ylab="running average (blue), datapoints (red)",ylim=c(-2,45))
points(result,col="blue")
```
